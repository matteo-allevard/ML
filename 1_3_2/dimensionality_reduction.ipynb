# Conclusion
Après avoir testé plusieurs méthodes de réduction de dimension (PCA et t-SNE), t-SNE en dimension 2 est celle qui sépare le mieux les deux classes.


```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

data = np.load("../data/dimensionality_reduction/data.npy")
labels = np.load("../data/dimensionality_reduction/labels.npy")

print("Shape des données :", data.shape)
print("Shape des labels :", labels.shape)

pca_2 = PCA(n_components=2)
data_pca_2 = pca_2.fit_transform(data)

pca_3 = PCA(n_components=3)
data_pca_3 = pca_3.fit_transform(data)

tsne_2 = TSNE(n_components=2, learning_rate='auto', init='pca', perplexity=30)
data_tsne_2 = tsne_2.fit_transform(data)

tsne_3 = TSNE(n_components=3, learning_rate='auto', init='pca', perplexity=30)
data_tsne_3 = tsne_3.fit_transform(data)

plt.figure(figsize=(6, 5))
plt.scatter(data_pca_2[:, 0], data_pca_2[:, 1], c=labels, cmap="coolwarm", s=10)
plt.title("PCA - Réduction en 2D")
plt.show()

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data_pca_3[:, 0], data_pca_3[:, 1], data_pca_3[:, 2], c=labels, cmap="coolwarm", s=10)
plt.title("PCA - Réduction en 3D")
plt.show()

plt.figure(figsize=(6, 5))
plt.scatter(data_tsne_2[:, 0], data_tsne_2[:, 1], c=labels, cmap="coolwarm", s=10)
plt.title("t-SNE - Réduction en 2D")
plt.show()

fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data_tsne_3[:, 0], data_tsne_3[:, 1], data_tsne_3[:, 2], c=labels, cmap="coolwarm", s=10)
plt.title("t-SNE - Réduction en 3D")
plt.show()
```
