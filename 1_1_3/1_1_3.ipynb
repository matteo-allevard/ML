1. Définition de la variable Z

```python
%pip install numpy matplotlib

import numpy as np # type: ignore
import matplotlib.pyplot as plt  # type: ignore

mu_X = 30
mu_Y = 15
sigma_X = 10
sigma_Y = 5

exp_Z = np.array([mu_X, mu_Y])
```

2. Génération de n et leur affichage

```python
N = 1000

X_samples = np.random.normal(mu_X, sigma_X, N)
Y_samples = np.random.normal(mu_Y, sigma_Y, N)

plt.figure(figsize=(6,5))
plt.scatter(X_samples, Y_samples, alpha=0.5)
plt.xlabel("Vitesse du vent (km/h)")
plt.ylabel("Température (°C)")
plt.title("Échantillons de Z")
plt.show()
```

3. Convergence moyenne empirique vers l'espérance

```python
distances = []


empirical_mean = np.zeros(2)


for n in range(1, N + 1):
	empirical_mean = np.array([np.mean(X_samples[:n]), np.mean(Y_samples[:n])])
	dist = np.linalg.norm(empirical_mean - exp_Z)
	distances.append(dist)

plt.figure(figsize=(7,5))
plt.plot(range(1, N+1), distances)
plt.xlabel("n (nombre d'échantillons)")
plt.ylabel("Distance à l'espérance")
plt.title("Convergence de la moyenne empirique vers l'espérance")
plt.show()
```
